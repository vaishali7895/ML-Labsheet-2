```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. Create a NumPy array from a Python list and print its shape and datatype.
py_list = [1, 2, 3, 4, 5]
arr = np.array(py_list)
print("Array:", arr)
print("Shape:", arr.shape)
print("Data type:", arr.dtype)
```

```python
# 2. Create a 2D array of size 4x4 filled with random numbers.
arr2d = np.random.rand(4, 4)
print("2D Random Array:\n", arr2d)
```

```python
# 3. Generate a NumPy array of 10 evenly spaced values between 1 and 100.
arr_linspace = np.linspace(1, 100, 10)
print("Evenly spaced values:\n", arr_linspace)
```

```python
# 4. Perform addition, subtraction, and multiplication on two NumPy arrays.
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
```

```python
# 5. Reshape a 1D NumPy array of 12 elements into a 3x4 matrix.
arr1d = np.arange(12)
reshaped = arr1d.reshape(3, 4)
print("Reshaped 3x4 Matrix:\n", reshaped)
```

```python
# 6. Calculate mean, median, and standard deviation of a NumPy array.
data = np.array([10, 20, 30, 40, 50])
print("Mean:", np.mean(data))
print("Median:", np.median(data))
print("Standard Deviation:", np.std(data))
```

```python
# 7. Use slicing to extract a subarray from a larger 2D NumPy array.
arr2d = np.arange(1, 17).reshape(4, 4)
print("Original Array:\n", arr2d)
subarray = arr2d[1:3, 1:3]
print("Subarray (2x2 block):\n", subarray)
```

```python
# 8. Save a NumPy array to a .npy file and reload it.
np.save("my_array.npy", arr2d)
loaded_arr = np.load("my_array.npy")
print("Reloaded Array:\n", loaded_arr)
```

```python
# 9. Demonstrate element-wise comparison operations in NumPy.
a = np.array([1, 2, 3, 4])
b = np.array([2, 2, 4, 4])
print("a == b:", a == b)
print("a > b:", a > b)
print("a < b:", a < b)
```

```python
# 10. Create an identity matrix using NumPy and display it.
identity = np.eye(5)
print("Identity Matrix:\n", identity)
```

```python
# 11. Create a Pandas DataFrame using a dictionary of lists.
data = {
    "Name": ["Priyanshu", "Aman", "Ravi", "Shushi"],
    "Age": [23, 25, 22, 24],
    "Marks": [99, 70, 85, 76]
}
df = pd.DataFrame(data)
print("DataFrame:\n", df)
```

```python
# 12. Filter rows in a DataFrame where a specific column value is greater than a threshold.
print("Filter Marks > 80:\n", df[df["Marks"] > 80])
```

```python
# 13. Add a new column to a DataFrame based on an existing columnâ€™s values.
df["Grade"] = ["A" if x > 85 else "B" for x in df["Marks"]]
print("DataFrame with Grade:\n", df)
```

```python
# 14. Group data by a categorical column and calculate the average of numeric columns.
grouped = df.groupby("Grade").mean(numeric_only=True)
print("Grouped by Grade:\n", grouped)
```

```python
# 15. Plot a line graph using Matplotlib for a dataset column over time.
time = pd.date_range(start="2025-01-01", periods=4, freq="M")
plt.plot(time, df["Marks"], marker='o', label="Marks")
plt.title("Marks Over Time")
plt.xlabel("Date")
plt.xticks(rotation=45)
plt.ylabel("Marks")
plt.legend()
plt.show()
```

```python
# 16. Create a bar chart showing frequency of categories in a dataset column.
df["Grade"].value_counts().plot(kind="bar")
plt.title("Frequency of Grades")
plt.xlabel("Grade")
plt.ylabel("Count")
plt.show()
```

```python
# 17. Generate a histogram of a numerical column from a dataset using Matplotlib.
plt.hist(df["Marks"], bins=6, edgecolor='black')
plt.title("Marks Distribution")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.show()
```

```python
# 18. Plot a scatter plot of two numerical variables using Matplotlib.
plt.scatter(df["Age"], df["Marks"], color="red")
plt.title("Age vs Marks")
plt.xlabel("Age")
plt.ylabel("Marks")
plt.show()
```

```python
# 19. Customize your plot: add title, axis labels, and legend using Matplotlib.
plt.plot(df["Age"], df["Marks"], label="Marks by Age", color="green", marker="o")
plt.title("Customized Plot")
plt.xlabel("Age")
plt.ylabel("Marks")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 20. Save a Matplotlib plot to a file (PNG/JPG) and display it inline in Jupyter Notebook.
plt.plot(df["Age"], df["Marks"], marker="o")
plt.title("Save Plot Example")
plt.savefig("plot.png")
plt.show()
```

```python
data = {
    "Name": ["Priyanshu", "Shahvez", "Ansh", "Shushi"],
    "Age": [23, 25, 22, 24],
    "Marks": [85, 90, 88, 76]
}
df = pd.DataFrame(data)
df["Grade"] = ["A" if x > 85 else "B" for x in df["Marks"]]
sns.set_theme(style="whitegrid", palette="bright")
```

```python
time = pd.date_range(start="2023-01-01", periods=4, freq="M")
df["Date"] = time
plt.figure(figsize=(7,4))
sns.lineplot(x="Date", y="Marks", data=df, marker="o", linewidth=2, color="purple")
plt.title("Marks Over Time", fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Grade", data=df, palette="Set2")
plt.title("frequency of Grades", fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(7,4))
sns.histplot(df["Marks"], bins=5, kde=True, color="orange")
plt.title("Marks Distribution", fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(7,4))
sns.scatterplot(x="Age", y="Marks", data=df, hue="Grade", size="Marks", palette="coolwarm", sizes=(50,200))
plt.title("Age vs Marks", fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(7,4))
sns.lineplot(x="Age", y="Marks", data=df, marker="o", color="green", label="Marks by Age")
plt.title("Customized Plot", fontsize=14)
plt.xlabel("Age")
plt.ylabel("Marks")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(7,4))
sns.lineplot(x="Age", y="Marks", data=df, marker="o", color="orange")
plt.title("Save Plot Example", fontsize=14)
plt.savefig("seaborn_plot.png")
plt.show()
```
